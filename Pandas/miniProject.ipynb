```python
import pandas as pd
```

```python
df = pd.read_csv('miniProject.csv')
print(df.head())
```

```python
print(df.info())
```

```python
print("Columns - ",list(df.columns))
print(f"dimension = {list(df.shape)}")
```

```python
# top revennue generating customers
customers = df.groupby("Customer Name")["Total"].sum().sort_values(ascending=False)
customers
```

```python
# average order value per customer
ans = df.groupby("Customer Name")["Total"].mean()
ans
```

```python
# total quantity ordered per category
ans = df.groupby("Category")['Quantity'].sum()
ans 
```

```python
# orders in south region with Price > 700
ans = df.query("Region == 'South' and Price > 700")
ans
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df
```

```python
df['Month'] = pd.to_datetime(df['Order Date']).dt.month_name()
df
```

```python
df['Month'] = pd.to_datetime(df['Order Date']).dt.to_period('Y')
df
```

```python
monthly_revenue  = df.groupby('Month')['Total'].sum().reset_index()
monthly_revenue
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(monthly_revenue['Month'], monthly_revenue['Total'], color='blue')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
