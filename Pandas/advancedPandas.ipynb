apply(), map(), applymap()

```python
# task 1
def set_size(qty):
    if qty >= 20:
        return 'Bulk'
    elif qty >= 15:
        return 'Medium'
    else:
        return 'Small'

df['Order Size'] = df['Quantity'].apply(set_size)
df.head(3)
```

```python
df['Price Category'] = df['Price'].map(lambda x: 'High' if x > 1000 else "Medium" if x > 500 else 'Low')
df.head(3)
```

```python
df['Rounded Total'] = df['Total'].apply(round)
df.head()
```

```python
df['Discount'] = df['Total'].map(lambda x: x * 0.10 )
df['Total SP'] = df['Total']- df['Discount']
df
```

agg(), transform(), filter()

```python
df['Customer Avg Total'] = df.groupby('Customer Name')['Total'].transform('mean')
df
```

```python
ans = df.groupby('Category').filter(lambda s: s['Quantity'].sum() > 10.0 )
ans.head()
```

```python
def set_order_size(q):
    if q >= 15:
        return "High"
    return "Low"

df['Order Size'] = df['Quantity'].apply(set_order_size)
df
```
