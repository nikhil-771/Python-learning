```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(0, 11, 5)
y = np.random.randint(10, 100, 5)
data = np.random.rand(1000)
```

```python
# bar chart
fig = plt.figure(figsize=(6,4))
plt.title("New iamge")
plt.bar(['a', 'b', 'c', 'd', 'e'], x)
plt.grid(True)
plt.legend(["data"])
fig.savefig('New fig')
plt.show()
```

```python
# pie chart
x = np.random.randint(0,11, 5)
plt.pie(x,   labels=['a','b', 'c', 'd' , 'e'] ,autopct='%1.2f%%')
plt.show()
```
