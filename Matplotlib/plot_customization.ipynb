```python
import matplotlib.pyplot as plt
```

```python

x = [1, 2, 3, 4, 5]
y = [30, 50, 40, 90, 70]
```

```python
plt.plot(x, y, 'g.-', label='Label')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("This is title")
plt.legend(loc='upper left')


plt.xlim(0, 6)
plt.ylim(10, 100)

plt.annotate("Highest", xy=(4, 90), xytext=(2.5, 95),
             arrowprops=dict(facecolor='black', arrowstyle="->"))

plt.show()
```
