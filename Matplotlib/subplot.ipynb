```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 18]
data = np.random.randn(100)

plt.figure(figsize=(10, 8))
```

```python
plt.subplot(2,2, 1)
plt.bar(x, y)
plt.title("Bar chart")

plt.subplot(2,2,2)
plt.scatter(x,y)
plt.title("Scatter Plot")

plt.subplot(2,2,3)
plt.hist(data, bins=10, edgecolor='black')
plt.title("Histogram")

plt.subplot(2,2,4)
plt.plot(x, y)
plt.title("Line Chart")

plt.tight_layout()
plt.show()
```
