```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
st = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
mat = [88, 76, 93, 85, 91]
cs = [78, 82, 89, 94, 90]
physics = [84, 79, 85, 87, 88]
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(st, mat, color='green', label='Maths', marker='.')
plt.plot(st, cs, color='blue', label='Computer Science', marker='o')
plt.plot(st, physics, color='brown', label='Physics', marker='+')
plt.legend()
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.title("Student marks", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))

x  = np.arange(len(st))
width = 0.25

plt.bar(x-width, mat, width=width, label='Math', color='blue')
plt.bar(x, cs, color='green', width=width, label='CS')
plt.bar(x+width, physics, label='Physics',  width= width,color='red')

plt.xticks(x, st)

plt.show()
```
