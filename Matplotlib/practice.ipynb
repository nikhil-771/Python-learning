```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(1000)
y = np.arange(100)
x = np.random.randint(100, 500, 100)
```

```python
fig = plt.figure(figsize=(6,4))
plt.plot(y,x, label='ye lable h', color='green', linewidth=1, marker='.')
plt.title("This is title", fontsize=15)
plt.xlabel("this is x")
plt.ylabel("This is y")
plt.grid(True, linestyle=':', alpha = 0.9)
plt.legend()
plt.annotate('Important Point', xy=(x[2], y[2]), xytext=(x[2]+1, y[2]+5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
plt.savefig("Line chart.png")
```

```python
fig = plt.figure(figsize=(6,4))
plt.scatter(x, y, label='Label', marker='.', color='blue')
plt.show()
```

```python
fig = plt.figure(figsize=(6,4))
plt.hist(np.random.rand(100), bins=10, color='purple', edgecolor='black')
plt.show()
```

```python
marks = [20, 35, 30]
sub = ['Math', 'CS', 'Physics']
plt.bar(sub, marks )
plt.show()
```

```python

scores = [20, 35, 30]
subjects = ['Math', 'CS', 'Physics']

# Plot
plt.bar(subjects, scores, color='skyblue')

# Title and axis labels
plt.title("Scores by Subject")
plt.xlabel("Subjects")
plt.ylabel("Scores")

# Show values on top of bars
for i, score in enumerate(scores):
    plt.text(i, score + 1, str(score))

# Display plot
plt.show()
```
