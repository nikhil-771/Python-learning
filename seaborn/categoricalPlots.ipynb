```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
var = sns.load_dataset('tips')
var.head()
```

```python
data = np.linspace(10,70, 7)
data = np.append(data, 120)
data
```

```python
# boxplot using matplotlib
plt.boxplot(data, vert=True, labels=['python'], showmeans=True)
plt.show()
```

```python
# boxplot using matplotlib
plt.figure(figsize=(6,5))
# sns.set(style="whitegrid")
sns.boxplot(x='day', y='total_bill', data=var, hue='sex',showmeans=True,palette='pastel' )
plt.tight_layout()
plt.show()
```

```python
# violin plot
sns.set(style='white')
sns.violinplot(x='day', y='total_bill', data=var, palette='cool')
plt.show()
```

```python
# Pair plot
sns.set(style='darkgrid')
sns.pairplot(var, hue='sex', kind='kde')
plt.plot()
```
