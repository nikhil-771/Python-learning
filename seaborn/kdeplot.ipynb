```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = tips = sns.load_dataset('tips')
data.head()
```

```python
sns.kdeplot(data=tips, x="total_bill", fill=True, color='green')
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='time', fill=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill', y='tip', levels=4)
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill', clip=(0, 50))
```
