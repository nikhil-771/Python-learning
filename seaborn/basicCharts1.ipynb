```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dt = tips = sns.load_dataset('tips')
print(dt.head(3))
```

```python
# scatter plot
sns.scatterplot(x='total_bill', y='tip', data=dt)
plt.xlabel("x label")
plt.show()
```

```python
# line plot
sns.lineplot(x='size', y='total_bill', data=dt)
plt.show()
```

```python
# bar plot
sns.barplot(x='day', y='total_bill', data=dt)
plt.show()
```

```python
# count plot
sns.countplot(x='day', data=tips)
plt.show()
```

```python
# box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
# violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```
