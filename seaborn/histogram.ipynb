```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = tips = sns.load_dataset('tips')
data.head()
```

```python
# Histogram
sns.histplot(x='total_bill', data=data)
plt.plot()
```

```python
sns.histplot(data=tips, x='total_bill', hue='time', multiple='stack')
plt.show()
```

```python
sns.histplot( x='total_bill', data=tips.head(20), hue='day', multiple='dodge')
```

```python
sns.histplot(data=tips, x='total_bill')
plt.xticks(ticks=range(0, 55, 5))
plt.yticks(ticks=range(0,51,3))
plt.xlabel("Total Bill")
plt.show()
```

Bivariate Histogram

```python
sns.histplot(data=tips.head(20), x='total_bill', y='tip', hue='sex')
plt.show()
```
