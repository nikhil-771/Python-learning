```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = tips = sns.load_dataset('tips')
data.head()
```

```python
sns.displot(data=tips, x="total_bill")  
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", kind='ecdf')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', col='time')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', col='time', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', kind='kde')
plt.show()
```

```python
sns.displot(
    data=tips,
    x='total_bill',
    hue='day',
    col='sex',
    kind='hist',
    stat='density',
    common_norm=False,
    height=4,
    facet_kws={'margin_titles': True}
)
plt.show()
```
