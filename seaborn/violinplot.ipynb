```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = tips = sns.load_dataset('tips')
data.head()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
sns.violinplot(data=tips, y='day', x='total_bill')
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', 
               hue='sex', split=True)
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', 
               inner=None)
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', 
               inner='stick')
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', 
               inner='quartile')
plt.show()
```
